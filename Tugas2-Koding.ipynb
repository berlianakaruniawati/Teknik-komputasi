Soal 1


Metode Bisection


```python
def bisection(f, a, b, tol=1e-6, max_iter=100):

    for i in range(max_iter):
        c = (a + b)/2

        if abs(f(c)) < tol:
            return c

        if f(a)*f(c) < 0:
            b = c
        else:
            a = c

    return c
```

Metode Regulafalsi

```python
def regulafalsi(f, a, b, tol=1e-6, max_iter=100):

    for i in range(max_iter):

        xr = b - (f(b)*(b-a))/(f(b)-f(a))

        if abs(f(xr)) < tol:
            return xr

        if f(a)*f(xr) < 0:
            b = xr
        else:
            a = xr

    return xr
```

Metode Secant

```python
def secant(f, x0, x1, tol=1e-6, max_iter=100):

    for i in range(max_iter):

        x2 = x1 - (f(x1)*(x1-x0))/(f(x1)-f(x0))

        if abs(f(x2)) < tol:
            return x2

        x0 = x1
        x1 = x2

    return x2
```

Metode Newtonrap

```python
def newtonrap(f, df, x0, tol=1e-6, max_iter=100):

    for i in range(max_iter):

        x1 = x0 - f(x0)/df(x0)

        if abs(f(x1)) < tol:
            return x1

        x0 = x1

    return x1
```

Metode Regulafalsi

```python
import numpy as np

def f(x):
    return x**2 - x - 2

def regulafalsi_table(f, a, b, tol=1e-6, max_iter=20):

    print("METODE REGULAFALSI")
    print("{:<8} {:<12} {:<12} {:<12} {:<12}".format("Iterasi","a","b","xr","f(xr)"))
    print("-"*60)

    for i in range(max_iter):

        xr = b - (f(b)*(b-a))/(f(b)-f(a))

        print("{:<8d} {:<12.6f} {:<12.6f} {:<12.6f} {:<12.6f}".format(i+1,a,b,xr,f(xr)))

        if abs(f(xr)) < tol:
            return xr

        if f(a)*f(xr) < 0:
            b = xr
        else:
            a = xr

    return xr


root_rf = regulafalsi_table(f,1,3)
print("\nAkar pendekatan:",root_rf)
```

Metode Secant

```python
import numpy as np

def f(x):
    return x**2 - x - 2

def secant_table(f, x0, x1, tol=1e-6, max_iter=20):

    print("METODE SECANT")
    print("{:<8} {:<12} {:<12} {:<12} {:<12}".format("Iterasi","x0","x1","x2","f(x2)"))
    print("-"*60)

    for i in range(max_iter):

        x2 = x1 - (f(x1)*(x1-x0))/(f(x1)-f(x0))

        print("{:<8d} {:<12.6f} {:<12.6f} {:<12.6f} {:<12.6f}".format(i+1,x0,x1,x2,f(x2)))

        if abs(f(x2)) < tol:
            return x2

        x0 = x1
        x1 = x2

    return x2


root_sec = secant_table(f,1,3)
print("\nAkar pendekatan:",root_sec)
```

Soal 4

Gerak vertikal.
Tinggi benda dapat ditulis dengan persamaan h = v0·t − 0.5·g·t². Misalnya kecepatan awal v0 = 20 m/s dan percepatan gravitasi g = 10 m/s², lalu ingin dicari waktu ketika tinggi benda mencapai 15 m. Persamaan tersebut diubah menjadi f(t) = 20t − 5t² − 15 = 0, kemudian nilai t dicari menggunakan metode numerik. (metode bisection)


```python
def f(t):
    return 20*t - 5*t**2 - 15

def bisection(f, a, b, tol=1e-6, max_iter=100):

    for i in range(max_iter):

        c = (a + b)/2

        if abs(f(c)) < tol:
            return c

        if f(a)*f(c) < 0:
            b = c
        else:
            a = c

    return c

t1 = bisection(f,0,2)
t2 = bisection(f,2,5)

print("Waktu pertama:", t1, "detik")
print("Waktu kedua:", t2, "detik")
```

Soal 5

Metode Fixed Point Iteration dipakai untuk mencari nilai x dari persamaan x² − x − 2 = 0. Persamaan tersebut diubah menjadi bentuk x = √(x + 2). Kemudian dipilih nilai awal x, lalu dihitung berulang sampai nilainya hampir tidak berubah lagi. Dari perhitungan tersebut diperoleh nilai x mendekati 2.


```python
import numpy as np

def g(x):
    return np.sqrt(x + 2)

def fixed_point(g, x0, tol=1e-6, max_iter=100):

    for i in range(max_iter):

        x1 = g(x0)

        if abs(x1 - x0) < tol:
            return x1

        x0 = x1

    return x1

root = fixed_point(g,1)

print("Akar pendekatan:",root)
```
