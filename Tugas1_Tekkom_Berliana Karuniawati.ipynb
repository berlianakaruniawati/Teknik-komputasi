```
sentence = "this is a rambling sentence that simply goes on and on and on and just simply will not stop that is just the way things are"

words = sentence.split()

wordtocount = {}

for word in words:
    wordtocount[word] = wordtocount.get(word, 0) + 1

print("Kata\tJumlah")
for word, count in wordtocount.items():
    print(f"{word}\t{count}")

once = 0
twice = 0
thrice = 0

for count in wordtocount.values():
    if count == 1:
        once += 1
    elif count == 2:
        twice += 1
    elif count == 3:
        thrice += 1

print("\nJumlah kata yang muncul sekali:", once)
print("Jumlah kata yang muncul dua kali:", twice)
print("Jumlah kata yang muncul tiga kali:", thrice)

total_words = len(words)
cek_total = once + 2*twice + 3*thrice

print("\nTotal kata sebenarnya:", total_words)
print("Total berdasarkan perhitungan:", cek_total)
```

```
def jumlah_ganjil(n):
    total = 0
    for i in range(1, n + 1, 2):
        total += i
    return total


while True:
    n = int(input("Masukkan bilangan ganjil (genap untuk berhenti): "))

    if n % 2 == 0:
        print("Program berhenti.")
        break

    hasil = jumlah_ganjil(n)
    print("Jumlah =", hasil)
```

```
def jumlah_ganjil(n):
    total = 0
    for i in range(1, n + 1, 2):
        total += i
    return total


with open("odd_in.dat", "w") as f:
    f.write("9\n15\n21\n33\n41\n")


with open("odd_in.dat", "r") as infile, open("odd_out.dat", "w") as outfile:
    counter = 1
    for line in infile:
        n = int(line.strip())
        hasil = jumlah_ganjil(n)
        outfile.write(f"{counter} {n} {hasil}\n")
        counter += 1


with open("odd_out.dat", "r") as f:
    print(f.read())
```

```
import math

with open("dummy1.dat", "w") as f:
    f.write("hello world 21.2 12.1\n")
    f.write("how do 1.7 123.5\n")
    f.write("you obfuscate 0.004 19.5\n")


total_sum = 0

with open("dummy1.dat", "r") as infile, open("dummy2.dat", "w") as outfile:
    counter = 1
    for line in infile:
        parts = line.split()
        num1 = float(parts[-2])
        num2 = float(parts[-1])

        total_sum += num1 + num2

        outfile.write(f"{counter} {num2} {math.exp(num2)}\n")
        counter += 1

    outfile.write(f"Total sum: {total_sum}\n")


with open("dummy2.dat", "r") as f:
    print(f.read())
```

```
def readtable(filename):
    xs = []
    ys = []

    with open(filename, "r") as f:
        for line in f:
            parts = line.split()
            xs.append(float(parts[0]))
            ys.append(float(parts[1]))

    return xs, ys
```

```
def readtable_lc(filename):
    with open(filename, "r") as f:
        data = [line.split() for line in f]

    xs = [float(row[0]) for row in data]
    ys = [float(row[1]) for row in data]

    return xs, ys
```

```
import math

def f(x):
    return math.exp(-x**4)

xs = [round(-1 + 0.1*i, 1) for i in range(21)]
ys = [f(x) for x in xs]

for x, y in zip(xs, ys):
    print(f"{x:.1f} {y:.5f}")
```
