Definisi masalah fisika:

Dalam suatu sistem mekanika terdapat beberapa gaya yang bekerja secara bersamaan pada sebuah benda. Besarnya gaya-gaya tersebut saling berhubungan sehingga dapat dinyatakan dalam bentuk sistem persamaan linear. Untuk menentukan nilai setiap gaya yang bekerja, digunakan metode numerik yaitu metode Gauss–Jordan dengan pivoting.

Rumusan matematis:

Hubungan gaya dapat dituliskan sebagai sistem persamaan linear:

x + y + z + w = 10
2x + y + z + w = 12
x + 2y + z + w = 13
x + y + 2z + w = 14

Sistem persamaan ini kemudian diselesaikan menggunakan metode Gauss-Jordan untuk mendapatkan nilai setiap variabel.

Implementasi Pmrograman:

```python
import numpy as np

def gauss_jordan(A, b):

    n = len(b)

    M = np.hstack((A, b.reshape(-1,1)))

    for k in range(n):

        max_row = np.argmax(abs(M[k:,k])) + k
        M[[k, max_row]] = M[[max_row, k]]

        M[k] = M[k] / M[k,k]

        for i in range(n):
            if i != k:
                M[i] = M[i] - M[i,k] * M[k]

    return M[:, -1]


A = np.array([
[1,1,1,1],
[2,1,1,1],
[1,2,1,1],
[1,1,2,1]
], dtype=float)

b = np.array([10,12,13,14], dtype=float)

solusi = gauss_jordan(A,b)

print("Solusi:")
print("x1 =", solusi[0])
print("x2 =", solusi[1])
print("x3 =", solusi[2])
print("x4 =", solusi[3])
```
