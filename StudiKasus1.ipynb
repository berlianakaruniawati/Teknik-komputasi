1. Pendahuluan

Atmosfer merupakan salah satu aspek penting dalam mempelajari kondisi suatu planet, termasuk Mars. Salah satu parameter utama dalam atmosfer adalah tekanan udara yang berubah terhadap ketinggian. Pada Mars, tekanan atmosfer jauh lebih rendah dibandingkan Bumi dan cenderung menurun secara signifikan seiring bertambahnya ketinggian. Hal ini disebabkan oleh pengaruh gravitasi serta distribusi suhu di atmosfer.

Mars dipilih sebagai objek studi karena memiliki atmosfer yang relatif sederhana dibandingkan planet gas raksasa, namun tetap cukup kompleks sehingga memerlukan pendekatan numerik dalam analisisnya. Selain itu, data atmosfer Mars tersedia secara luas dari berbagai misi eksplorasi. Dibandingkan Bumi yang sering dapat diselesaikan secara analitik, Mars memberikan kasus yang lebih sesuai untuk penerapan metode numerik seperti Simpson.

Dalam perhitungan tekanan atmosfer, biasanya digunakan persamaan fisika yang melibatkan hubungan antara tekanan, massa jenis, dan suhu. Namun, ketika suhu berubah terhadap ketinggian dalam bentuk fungsi non-linear, penyelesaian secara analitik menjadi sulit dilakukan. Oleh karena itu, diperlukan metode numerik untuk mendekati solusi dari permasalahan tersebut. Pada studi kasus ini, digunakan metode Simpson untuk menghitung tekanan atmosfer Mars pada ketinggian tertentu.

2. Metodologi

Perubahan tekanan terhadap ketinggian mengikuti persamaan hidrostatik, yaitu perubahan tekanan sebanding dengan massa jenis fluida dan percepatan gravitasi. Dengan menggabungkan persamaan tersebut dengan hukum gas ideal, diperoleh hubungan yang menyatakan bahwa perubahan tekanan bergantung pada suhu sebagai fungsi ketinggian. Dari proses ini, didapatkan bentuk persamaan yang mengandung integral terhadap fungsi suhu.

Persamaan akhir menunjukkan bahwa tekanan pada ketinggian tertentu bergantung pada hasil integral dari fungsi yang melibatkan suhu. Namun, karena fungsi suhu yang digunakan berbentuk non-linear (mengandung kombinasi fungsi linear dan eksponensial), integral tersebut tidak dapat diselesaikan secara analitik. Oleh karena itu, digunakan metode numerik.

Metode numerik yang digunakan dalam studi kasus ini adalah metode Simpson, yang termasuk dalam kategori integral numerik. Metode ini bekerja dengan membagi interval menjadi beberapa bagian kecil, kemudian mendekati luas kurva menggunakan polinomial orde dua. Dibandingkan metode lain seperti trapezoidal, Simpson memiliki tingkat akurasi yang lebih baik untuk fungsi yang halus seperti profil suhu atmosfer.
